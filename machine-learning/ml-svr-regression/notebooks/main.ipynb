```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sb

from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv('./assets/raw/SP5_Abalone.csv')
```

```python
df = df.iloc[:,1:]
```

```python
sb.pairplot(df, y_vars='Rings')
```

```python
df.columns
```

```python
cols = ['Length', 'Diameter', 'Height', 'Whole weight', 'Shucked weight', 'Viscera weight', 'Shell weight', 'Rings']
```

```python
x_values = df[cols].values
y_values = df['Rings'].values

x_train, x_test, y_train, y_test = train_test_split(x_values, y_values, test_size=0.2, random_state=42)
```

```python
svr = SVR(kernel='rbf')

svr.fit(x_train, y_train)
```

```python
predicts = svr.predict(x_test)
r2 = r2_score(y_test, predicts)
mse = mean_squared_error(y_test, predicts)

print("Predictions: {}, Current: {}".format(predicts[:4], y_test[:4]))
print("Mean Squared Error: ", mse)
print("R2 Score: ", r2)

plt.figure(figsize=(10, 6))
sb.regplot(x=y_test, y=predicts, scatter_kws={'color': 'blue'}, line_kws={'color': 'red'}, ci=None)

plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)

plt.title('Comparacin entre Predicciones y Valores Reales (SVR)')
plt.xlabel('Current Values')
plt.ylabel('Predictions')
plt.show()
```
