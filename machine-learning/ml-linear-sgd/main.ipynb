```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import fetch_openml
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
```

```python
MNIST = fetch_openml('mnist_784', version=1)
```

```python
MNIST.keys()
```

```python
x_values = MNIST['data']
y_values = MNIST['target']

print(x_values.shape)
print(y_values.shape)
```

```python
digit = x_values.to_numpy()[7]
digit_img = digit.reshape(28,28)

plt.imshow(digit_img, cmap='binary')
plt.show()
```

```python
y_values = y_values.astype(np.uint8)

y_values[7]
```

```python
x_train, x_test, y_train, y_test = x_values[:60000], x_values[60000:], y_values[:60000], y_values[60000:]

y_train_three = (y_train == 3)
y_test_three = (y_test == 3)
```

```python
sgd = SGDClassifier(random_state=42)

sgd.fit(x_train, y_train_three)
```

```python
predicts = sgd.predict(x_test)
```

```python
accuracy = accuracy_score(y_test_three, predicts)
conf_matrix = confusion_matrix(y_test_three, predicts)
precision = precision_score(y_test_three, predicts)
recall = recall_score(y_test_three, predicts)
f1 = f1_score(y_test_three, predicts)
```

```python
print("Accuracy Score: ", accuracy)
print("Confussion Matrix: ")
print("")
print(conf_matrix)
print("")
print("Recall Score: ", recall)
print("Precision Score: ", precision)
print("F1 Score:", f1)
```
