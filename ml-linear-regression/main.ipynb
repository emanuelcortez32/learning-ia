```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv('./assets/raw/SP5_Abalone.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.columns
```

```python
cols = ['Sex', 'Length', 'Diameter', 'Height', 'Whole weight', 'Shucked weight', 'Viscera weight', 'Shell weight']
```

```python
for col in cols:
    plt.plot(df[col], df['Rings'], 'ro')
    plt.title('Rings against at %s' % col)
    plt.ylabel('Rings')
    plt.xlabel(col)
    plt.show()
```

# Linear Regression Simple

```python
def simple_regression_lineal(x_independent):
    y_values = df['Rings'].values

    x_values = df[x_independent].values.reshape(-1,1)

    x_train, x_test, y_train, y_test = train_test_split(x_values, y_values, test_size=0.30, random_state=42)

    lin_reg = LinearRegression()
    lin_reg.fit(x_train, y_train)

    predicts = lin_reg.predict(x_test)

    mse = mean_squared_error(y_test, predicts)
    r2 = r2_score(y_test, predicts)

    print("Predicts: {}, Current: {}".format(predicts[:4], y_test[:4]))
    print("Mean Square Error: ", mse)
    print("R2 Score: ", r2)

    plt.title("Predictions based on %s" % col)
    plt.plot(x_test, y_test, 'ro')
    plt.plot(x_test, predicts)
    plt.show()
```

```python
for col in cols:
    simple_regression_lineal(col)
```

# Lineal Regression Multiple

```python
def multiple_lineal_regression(x_independents):
    y_values = df['Rings'].values

    x_values = df.loc[:, x_independents].values

    x_train, x_test, y_train, y_test = train_test_split(x_values, y_values, test_size=0.30, random_state=42)

    lin_reg = LinearRegression()
    lin_reg.fit(x_train, y_train)

    predicts = lin_reg.predict(x_test)

    mse = mean_squared_error(y_test, predicts)
    r2 = r2_score(y_test, predicts)

    print("Predicts: {}, Current: {}".format(predicts[:4], y_test[:4]))
    print("Mean Square Error: ", mse)
    print("R2 Score: ", r2)

    sb.regplot(x = y_test, y = predicts)
```

```python
multiple_lineal_regression(cols)
```
